En general, la idea es que la validación cruzada reduce la cantidad de errores que encontramos. Este efecto es fácilmente observable en Tic-tac-toe.
No podemos verlo en German, puesto que lla cantidad de errores que está cometiendo el algoritmo es tan pequeña que no parece dar ningún error en la mayoría de ejecuciones. Con más ejecuciones y más datos, observaríamos el mismo efecto que en tic-tac-toe.

El objetivo de usar Laplace, por otro lado, es el de suavizar el algoritmo al encontrarnos probabilidad 0, sumándole una constante alpha. Esto hace que nuestro Naive-Bayes se ajuste mejor a la realidad en caso de que nuestro *dataset* no tenga en cuenta todos los casos reales, y asigne por tanto una probabilidad de 0%.

### Naive-Bayes con Scikit-Learn

Este algoritmo se ha implementado en el archivo Scikit.py

```python
from abc import ABCMeta, abstractmethod
import numpy as np
from Datos import Datos

from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer 
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit

# https://github.com/scikit-learn/scikit-learn/blob/36958fb24/sklearn/naive_bayes.py#L750
"""
    Naive Bayes classifier for multinomial models.
    The multinomial Naive Bayes classifier is suitable for classification with
    discrete features (e.g., word counts for text classification). The
    multinomial distribution normally requires integer feature counts. However,
    in practice, fractional counts such as tf-idf may also work.
    Read more in the :ref:`User Guide <multinomial_naive_bayes>`.
"""
dataset = Datos("ConjuntosDatosIntroFAA/german.csv")
features = dataset.nominalAtributos[:-1]

ct = ColumnTransformer(dataset.nominalAtributos[:-1])

validacionSimple = ShuffleSplit(len(dataset.datos), test_size=.2, random_state=0)
validacionCruzada = 5

# Validacion Simple, sin Laplace
atributos = preprocessing.OneHotEncoder(sparse = False)
X = atributos.fit_transform(dataset.datos[:, :-1])
Y = dataset.datos[:, -1]

clasificador = MultinomialNB(alpha = 0.1, class_prior = None, fit_prior = True) # Alpha 0 -> No Laplace
clasificador.fit(X, Y)

valSimple = cross_val_score(clasificador, X, Y, cv = validacionSimple)

print("Validacion Simple sin usar Laplace:")
print("\tMedia: " + str(1 - np.mean(valSimple)))
print("\tVarianza: " + str(np.var(valSimple)) + "\n")
print("\n")

# Validacion Simple, con Laplace
atributos = preprocessing.OneHotEncoder(sparse = False)

X = atributos.fit_transform(dataset.datos[:, :-1])
Y = dataset.datos[:, -1]

clasificador = MultinomialNB(alpha = 1, class_prior = None, fit_prior = True) # Alpha 1 -> Laplace
clasificador.fit(X, Y)

valSimple = cross_val_score(clasificador, X, Y, cv = validacionSimple)

print("Validacion Simple usando Laplace:")
print("\tMedia: " + str(1 - np.mean(valSimple)))
print("\tVarianza: " + str(np.var(valSimple)) + "\n")
print("\n")

# Validacion Cruzada, sin Laplace
atributos = preprocessing.OneHotEncoder(sparse = False)
X = atributos.fit_transform(dataset.datos[:, :-1])
Y = dataset.datos[:, -1]

clasificador = MultinomialNB(alpha = 0.1, class_prior = None, fit_prior = True) # Alpha 0 -> No Laplace
clasificador.fit(X, Y)

valCruzada = cross_val_score(clasificador, X, Y, cv = validacionCruzada)

print("Validacion Cruzada sin usar Laplace:")
print("\tMedia: " + str(1 - np.mean(valCruzada)))
print("\tVarianza: " + str(np.var(valCruzada)) + "\n")
print("\n")

# Validacion Cruzada, con Laplace
atributos = preprocessing.OneHotEncoder(sparse = False)
X = atributos.fit_transform(dataset.datos[:, :-1])
Y = dataset.datos[:, -1]

clasificador = MultinomialNB(alpha = 1, class_prior = None, fit_prior = True) # Alpha 1 -> Laplace
clasificador.fit(X, Y)

valCruzada = cross_val_score(clasificador, X, Y, cv = validacionCruzada)

print("Validacion Cruzada usando Laplace:")
print("\tMedia: " + str(1 - np.mean(valCruzada)))
print("\tVarianza: " + str(np.var(valCruzada)) + "\n")
```
